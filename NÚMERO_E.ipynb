```
# EJERCICIO: 
# "Dada la siguiente función f(x)=x*e^(3*x), calcular el punto (x1,y1) donde se hace un mínimo en la funición, y representalo todo".
# NOTA: Para calcular el mínimo de una funcion, primero hay que caluclar su derivada y obtener su raiz d(x)=0, con este valor se 
# calcula el punto (x1,y1)

# 1º) Declaramos las librerías que vamos a utilizar para este ejercicio.
from math import *
from numpy import * 
from sympy import * 

# 2º)Definimos la función f(x), es decir la y (nota, aquí el número "e" se simboliza como "exp()") 
y=x*exp(3*x)

# 3º) Calculamos su derivada x = Symbol("x") # Para trabajar con Sympy hay que definir la X como un símbolo/variable a calcular 
# previamente indicamos a la librería de sympy cual es el símbolo de referencia
x=Symbol("x")  # ojo con la primer mayúscula
d= Derivative(y,x).doit()
print(d)
```

```
# 4º) Una vez conocida la función derivada d(x) obtenemos su raiz 
raiz= solve(d,x)
print(raiz)
```

```
# 5º) Obtenida la raiz, ya sabemos que en ese valor tenemos un mínimo. Ahora obtenemos el valor de la "y" en ese punto.
x1=round(-1/3,4)  # redondeo de x1 con 4 decimales
y1=round(x1*exp(3*x1),4)  # obtenemos el valor y1 redondeado a 4 decimales
print(x1,y1)
print(" El punto donde hay un MÍNIMO está en (x1,y1)=: (%.1f,%.1f)" %(x1,y1))
```

```
# Y por último vamos a representarlo todo,para ello declaramos las siguientes librerías
from math import *
from numpy import * 
from matplotlib.pyplot import * 

# primero vamos a declarar el segmento de representación del eje de abcisas "X"
x=linspace(-1,-0.15,200)

# y definimos las dos funciones a representar "y(x)" y su derivada "d(x)"
y=x*exp(3*x)
d=3*x*exp(3*x) + exp(3*x)

# Ahora definimos el tipo de ejes cartesianos que queremos usar
ax=subplot()
ax.spines['left'].set_position('zero')
ax.spines['bottom'].set_position('zero')


# Y ahora ploteamos la dos funciones en un mismo diagrama cartesiano
plot(x, y, x, d)

title("Gráfica $f(x)=3*exp(3*x)$") # Ponemos un título al diagrrama usando el formato LaTEX

scatter(x1, y1,label="MÍNIMO", color="red") # recalcamos el punto mínimo en la curva 
legend(loc="best") #colocación las ETIQUETAS en el mejor sítio que el programa disponga

# ponemos una flecha indicativa
annotate((x1,y1), xy=(x1, y1), xytext=(-0.7,0.1 ), arrowprops=dict(facecolor='red'))
r=[x1,x1]
s=[-0.1,0.15]
plot(r,s)
grid() # de forma opcional ponemos un enrejillado
show()  # y lo representamos todo
```
